```python
import pandas as pd
```

```python
layoff_data1 = "layoffs.csv"
```

```python
df1 = pd.read_csv(layoff_data1)
df1
```

```python
# Remove column name 'percentage_laid_off'
df1 = df1.drop(['percentage_laid_off','stage','funds_raised'], axis=1)
```

```python
df1
```

```python
layoff_data2 = "Tech_Layoffs_of_2022.csv"
df2 = pd.read_csv(layoff_data2)
df2
```

```python
# Remove column name 'percentage_laid_off'
df2 = df2.drop(['founded','month','year','m'], axis=1)
df2
```

```python
frames = [df1, df2]

result = pd.concat(frames)
result
```

```python
df_summary = result.groupby(['company','location','industry','country','date'], as_index=False).sum()
df_summary
```

```python
df_summary.to_csv('cleanlayoffs.csv')
```
